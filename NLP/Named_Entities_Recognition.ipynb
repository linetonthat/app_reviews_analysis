```python
# Quick function to remove ents formed on whitespace:
def remove_whitespace_entities(doc):
    doc.ents = [e for e in doc.ents if not e.text.isspace()]
    return doc

# Insert this into the pipeline AFTER the ner component:
nlp.add_pipe(remove_whitespace_entities, after='ner')
```

```python
for i in range(len(df)):
    doc = nlp(df.iloc[i]['review'])
    for sent in doc.sents:
        #docx = nlp(sent.text)
        docx = remove_whitespace_entities(nlp(sent.text))
        if docx.ents:
            #displacy.serve(docx, style = 'ent', options = options)
            #displacy.render(docx, style = 'ent', jupyter = True, options = options)
            displacy.render(docx, style = 'ent', jupyter = True)
        else:
            print(i, docx.text)
```

## Create columns for the locations recognised

```python
# add empty columns for GPE, LOC and NORP info
new_columns = list(df.columns) + ['GPE','LOC','NORP']
df = df.reindex(columns = new_columns)
df.head()
```

## Define functions to extract locations

```python
# define a function to collect list of entities from a string for a specific entity label (eg. 'GPE','LOC'...)
def has_specific_ent(s, label):
    get_ent_list = lambda ents: [ent.text for ent in ents if ent.label_ == label] 
    doc = nlp(s)
    # test if there are any named entities in the doc
    if doc.ents:
        # test if the list of named entity is not empty
        if len(get_ent_list(doc.ents)) >0:
            return get_ent_list(doc.ents)
```

```python
for l in ['GPE','LOC','NORP']:
    df[l] = df['review'].apply(has_specific_ent,label = l)
```

## Identify locations

```python
df_test = df.iloc[:60]
```

```python
import operator
from functools import reduce 
from collections import Counter
```

```python
# Define a function to retrieve list of locations from a dataset according to entity label
get_location_list = lambda dataset, label: list(dataset.loc[dataset[label].notnull(),label])
```

```python
# define a function to provide the count the occurences of locations
def count_locations(dataset, label):
    lst = get_location_list(dataset, label)
    concatenated_lst = reduce(operator.add, lst)
    return Counter(concatenated_lst).most_common()
```

```python
count_locations(df, 'GPE')
```

A large share of the locations are in the U.S., especially in California (California, CA, Bay area, San Francisco, SF, Los Angeles, LA). Asia appears as well, especially China (including Beijing) and South Korea (including Seoul).
